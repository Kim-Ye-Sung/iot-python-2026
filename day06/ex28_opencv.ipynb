## OpenCV 영상처리

### 웹캠 영상처리

```python
# 라이브러리 로드
import cv2
import numpy as np
import time
```

#### 기본 출력

```python
# 웹캠디바이스 연결하는데 시간소요
cap = cv2.VideoCapture(0)  # 첫번째 웹캠(0)

while True:
    ret, img = cap.read()   # 웹캠 화면은 하나씩 로드, ret(리턴결과), img(실제 이미지)
    if not ret : break

    cv2.imshow('webcam', img)
    if cv2.waitKey(1) == ord('q'):break     # ord('A') = 65, ord('a') = 97,  q(113) 키를 누르면 반복 종료

cap.release()   # 메모리 정리
cv2.destroyAllWindows() 
```

#### FPS(초당 프레임 수) 표시
- 대부분 30 FPS 사용함

```python
cap = cv2.VideoCapture(0)  # 첫번째 웹캠(0)

prev_time = 0  # 이전 프레임시간

while True:
    ret, img = cap.read()
    if not ret : break

    curr_time = time.time()  # 프레임 출력시간
    # FPS 계산
    fps = 1 /(curr_time - prev_time) if prev_time != 0 else 0
    prev_time = curr_time

    # 출력할 텍스트
    fps_text = f'FPS : {int(fps)}'  # int() 사용 안하면 소수점으로 출력됨
    # img 위에 텍스트 붙이기
    cv2.putText(img, fps_text, (10,30), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)

    cv2.imshow('webcam', img)
    if cv2.waitKey(1) == ord('q'):break

cap.release() 
cv2.destroyAllWindows() 
```

#### 해상도 조정
- 웹캠이 지원하는 해상도로 변경 가능

```python
cap = cv2.VideoCapture(0)  # 첫번째 웹캠(0)

if not cap.isOpened(): # 웹캠 오픈실패라면
    raise RuntimeError('웹캠 오픈 실패!')

# 해상도 설정
WIDTH, HEIGHT = 640, 480 
cap.set(cv2.CAP_PROP_FRAME_WIDTH, WIDTH)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, HEIGHT)

# 적용된 해상값 확인
real_w = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
real_h = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

prev_time = 0 

while True:
    ret, img = cap.read()
    if not ret : break

    curr_time = time.time() 
    
    fps = 1 /(curr_time - prev_time) if prev_time != 0 else 0
    prev_time = curr_time

    
    fps_text = f'FPS : {int(fps)}'  
    res_text = f'{int(real_w)}x{int(real_h)}'
    
    cv2.putText(img, fps_text, (10,30), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)
    cv2.putText(img, res_text, (10,70), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)

    cv2.imshow('webcam', img)
    
    key = cv2.waitKey(1)  # 키입력 변수
    if key == ord('q'):
        break
    elif key == ord('s'):  # 스냅샷 저장
        filename = time.strftime('snapshot_%Y%m%d%H%M%S.jpg')  # ./day06/ 밑에 저장됨
        cv2.imwrite(filename, img)
        print(f'{filename} 스냅샷 저장')

cap.release() 
cv2.destroyAllWindows()
```

#### 얼굴 검출
- OpenCV 내장된 기능
    - haarcascade_frontalface_default.xml
    - haarcascade_eye_tree_eyeglasses.xml
    - haarcascade_eye.xml
    - haarcascade_smile.xml
    - haarcascade_frontalcatface_extended.xml

```python
# Haar Cascade 파일 로드
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')  # 얼굴데이터 따로

smile_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_smile.xml')   # 미소데이터 따로

cap = cv2.VideoCapture(0)  # 첫번째 웹캠(0)

if not cap.isOpened(): 
    raise RuntimeError('웹캠 오픈 실패!')

real_w = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
real_h = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

prev_time = 0 
show_info = False 

while True:
    ret, img = cap.read()
    if not ret : break

    # FPS
    curr_time = time.time() 
    
    fps = 1 /(curr_time - prev_time) if prev_time != 0 else 0
    prev_time = curr_time

    if show_info:
        fps_text = f'FPS : {int(fps)}'  
        res_text = f'{int(real_w)}x{int(real_h)}'
    
        cv2.putText(img, fps_text, (10,30), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)
        cv2.putText(img, res_text, (10,70), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)
    
    # 물체검출
    faces = face_cascade.detectMultiScale(img, 1.1, 8)

    # 검출된 물체를 표시
    for(x, y, w, h) in faces:  # x, y 좌측상단좌표, w, h 너비와 높이
        cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,255), 2)

        # 얼굴 영역만 잘라서 ROI(Region of Interact)
        face_roi = img[y:y+h, x:x+w]

        # 얼굴 내부에서 미소 검출
        smiles = smile_cascade.detectMultiScale(face_roi, 1.7, 20)

        for(sx,sy,sw,sh) in smiles:
            cv2.rectangle(face_roi, (sx,sy), (sx+sw, sy+sh), (0,255,0),2)

    cv2.imshow('webcam', img)
    
    key = cv2.waitKey(1) 
    if key == ord('q'):
        break
    elif key == ord('s'): 
        filename = time.strftime('snapshot_%Y%m%d%H%M%S.jpg') 
        cv2.imwrite(filename, img)
        # print(f'{filename} 스냅샷 저장')
    elif key == 9:  
        show_info = not show_info  
        # print(f'표시여부 {show_info}')

cap.release() 
cv2.destroyAllWindows()
```

#### 얼굴 모자이크

```python
# Haar Cascade 파일 로드
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')  # 얼굴데이터 따로


cap = cv2.VideoCapture(0)  # 첫번째 웹캠(0)

if not cap.isOpened(): 
    raise RuntimeError('웹캠 오픈 실패!')

real_w = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
real_h = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)

prev_time = 0 
show_info = False 

while True:
    ret, img = cap.read()
    if not ret : break

    # FPS
    curr_time = time.time() 
    
    fps = 1 /(curr_time - prev_time) if prev_time != 0 else 0
    prev_time = curr_time

    if show_info:
        fps_text = f'FPS : {int(fps)}'  
        res_text = f'{int(real_w)}x{int(real_h)}'
    
        cv2.putText(img, fps_text, (10,30), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)
        cv2.putText(img, res_text, (10,70), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)
    
    # 물체검출
    faces = face_cascade.detectMultiScale(img, 1.1, 8)

    # 검출된 물체를 표시
    for(x, y, w, h) in faces:  # x, y 좌측상단좌표, w, h 너비와 높이
        cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,255), 2)

        # 얼굴 영역만 잘라서 ROI(Region of Interact)
        face_roi = img[y:y+h, x:x+w]

        # 모자이크 처리
        small = cv2.resize(face_roi, (w/15, h/15),interpolation=cv2.INTER_LINEAR)  # 작게 줄이고
        mosaic = cv2.resize(small, (w,h), interpolation=cv2.INTER_NEAREST)

        img[y:y+h, x:x+w] = mosaic  # ROI 교체


    cv2.imshow('webcam', img)
    
    key = cv2.waitKey(1) 
    if key == ord('q'):
        break
    elif key == ord('s'): 
        filename = time.strftime('snapshot_%Y%m%d%H%M%S.jpg') 
        cv2.imwrite(filename, img)
        # print(f'{filename} 스냅샷 저장')
    elif key == 9:  
        show_info = not show_info  
        # print(f'표시여부 {show_info}')

cap.release() 
cv2.destroyAllWindows()
```

#### QR코드 리더

- pyzbar 라이브러리 설치 - 구식 라이브러리로 호환 문제 발생, zbar 기초 라이브러리 문제 발생가능

```python
import webbrowser  # 웹브라우저용 모듈 로드

# QR코드 검출기 객체 생성
qr_detector = cv2.QRCodeDetector()

# 웹캠  열기
cap = cv2.VideoCapture(0)

is_opened = False  # 웹사이트 오픈 여부
last_data = ''   # 최종데이터

while True:
    ret, img = cap.read()
    if not ret:
        break
    
    # QR코드 검출 및 디코딩
    data, points, _ = qr_detector.detectAndDecode(img)

    # QR코드가 검출 되었으면
    if points is not None and data:
        points = points.astype(int)

        # QR코드 외곽선 그리기
        for i in range(len(points[0])):
            pt1 = tuple(points[0][i])
            pt2 = tuple(points[0][(i+1) % len(points[0])])
            cv2.line(img, pt1, pt2, (0,255,0), 2)


        # QR코드 내용 출력
        cv2.putText(img, data, (10,30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0,255,0),2)
        # print(data)  # 불필요한 로그

        # URL이고 아직 안열었으면 실행
        if(data.startswith('http://') or data.startswith('https://')):
            if not is_opened or last_data != data:  # 이미 열었거나, 데이터가 같으면 실행안함
                webbrowser.open(data)
                is_opened = True
                last_data = data
    
    else:  # QR코드 검출이 끝나면, QR코드가 화면에서 사라지면
        is_opened = False
        last_data = ''

    cv2.imshow("QR Code Reader", img)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
# 프레임 수마다 검사

import webbrowser  # 웹브라우저용 모듈 로드

# QR코드 검출기 객체 생성
qr_detector = cv2.QRCodeDetector()

# 웹캠  열기
cap = cv2.VideoCapture(0)

is_opened = False  # 웹사이트 오픈 여부
last_data = ''   # 최종데이터

frame_count = 0  # 프레임 수
det_interval = 30  # 10프레임마다 검사

while True:
    ret, img = cap.read()
    if not ret:
        break
    
    frame_count += 1
    data = ''
    points = None



    # 10프레임마다 검사
    if frame_count % det_interval == 0:
        data, points, _ = qr_detector.detectAndDecode(img)

    # QR코드가 검출 되었으면
    if points is not None and data:
        points = points.astype(int)

        # QR코드 외곽선 그리기
        for i in range(len(points[0])):
            pt1 = tuple(points[0][i])
            pt2 = tuple(points[0][(i+1) % len(points[0])])
            cv2.line(img, pt1, pt2, (0,255,0), 2)


        # QR코드 내용 출력
        cv2.putText(img, data, (10,30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0,255,0),2)
        # print(data)  # 불필요한 로그

        # 같은 QR 반복실행 방지
        if data != last_data or is_opened != True:
            if(data.startswith('http://') or data.startswith('https://')):
                webbrowser.open(data)
                is_opened = True
                last_data = data
    
    else:  # QR코드 검출이 끝나면, QR코드가 화면에서 사라지면
        is_opened = False
        last_data = ''

    cv2.imshow("QR Code Reader", img)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
